```python
# meta file list len : 
metas = []
meta_f_len = 0

dir_path = os.path.join(base_path, 'meta')

for (path, dir, file) in os.walk(dir_path):
#     print(path, dir, len(file))
    for f in file:
        metas.append(os.path.join(path, f))
#         print(os.path.join(path, f))
        meta_f_len += 1
print(meta_f_len, len(metas))
```

```python
meta_df = pd.DataFrame()
# cnt = 0 

for m in metas:
    with open(m, 'rt', encoding='UTF8') as j:
        meta_json = json.load(j)
#     print(meta_json['data_key'], meta_json['label_id'], meta_json['label_path'][0])
    meta_row = pd.DataFrame({'meta_path':m, 'meta_key':meta_json['data_key'], 'meta_label_id':meta_json['label_id'], 'meta_label_path': meta_json['label_path'][0]}, index=[0])
    meta_df = pd.concat([meta_df, meta_row])
    # cnt += 1
    # print(cnt)
    
meta_df.reset_index(drop=True, inplace=True)
meta_df
```

```python
def exctract_metainfo(df, colname, con1, con2):
    df[colname] = np.NaN

    for i, row in meta_df.iterrows():
    #     print(row['meta_path'])

        with open(row['meta_path'], 'rt', encoding='UTF8') as j:
            meta_json=json.load(j)    

        tags = [j['name'] for j in meta_json['tags']]
    #     print(tags, '\n')

        if con1 in tags:
            df[colname][i] = con1

        elif con2 in tags:
            df[colname][i] = con2
```

```python
# extract grade
def exctract_grade(json_data):
    dict_grade = dict()

    for j in json_data['properties'][:-1]:
        dict_grade[j['property_name']] = j['option_name']
    # print(dict_coord, dict_grade, '\n')
    
    return dict_grade
```

```python
label_info = pd.DataFrame()

for i, row in meta_df.iterrows():
    if row['Etday']=='D5' and row['type']=='microscope': #condition : D5 microscope
        base_path = row['meta_path'].split('meta')[0]
        label_path = os.path.join(base_path, row['meta_label_path'])
        
        with open(label_path, 'rt', encoding='UTF8') as j:
            label_json = json.load(j)
        
        if len(label_json) > 0: #not empty json file
#             print(i, row['meta_key'], label_path.split('/')[-1], len(label_json['objects']))
            
#             class_names= []
#             for o in label_json['objects']:
#                 class_names.append(o['class_name'])
#             print(class_names)

            grade = []
            box_coords = []
            for o in label_json['objects']:
                if o['class_name'] == 'fertilized_egg':
                    if exctract_grade(o) == {}:
#                         print('no info : ', row['meta_path'].split('/')[7:], row['Etday'], row['type'], label_path.split('/')[-2:], exctract_grade(o))
                        pass
                    else:
                        grades = exctract_grade(o) #exctract grade
                        box_coord = o['annotation']['coord'] #exctract bbox coord
                        grade.append(grades)
                        box_coords.append(box_coord)

#             print(row['meta_key'], label_path, exctract_grade(o), box_coord, '\n')

            meta_info = row
            meta_info['grades'] = grade
            meta_info['box_coord'] = box_coords
            label_row = meta_info.to_frame().transpose()

            label_info = pd.concat([label_info, label_row])
```

```python
def draw_bbox(img, box_coords):
    for coord in box_coords:
       
        point1 = [coord['x'], coord['y']]
        point2 =[coord['x'] + coord['width'], coord['y'] + coord['height']]
        # print(coord , '=>' ,tuple(map(int, point1)), tuple(map(int, point2)))

        cv2.rectangle(img, tuple(map(int, point1)), tuple(map(int, point2)), (0,0,0), 4)

def draw_ZONA(img, mask, label_json_object):
    one_list = []
    for o in label_json['objects']:
        if o['class_name'] == 'zona_outer':
            zona_outer_coords = o['annotation']['coord']['points'][0][0]
            points = []
            for k in range(len(zona_outer_coords)):
                points.append([int(zona_outer_coords[k]['x']), int(zona_outer_coords[k]['y'])])
            input_points = np.array(points, np.int32)
            one_list.append(input_points)


        elif o['class_name'] == 'zona_inner':
            zona_inner_coords = o['annotation']['coord']['points'][0][0]
            points = []
            for k in range(len(zona_inner_coords)):
                points.append([int(zona_inner_coords[k]['x']), int(zona_inner_coords[k]['y'])])
            input_points = np.array(points, np.int32)
            one_list.append(input_points)
            
    cv2.fillPoly(img, one_list, (0,255,0))
    cv2.fillPoly(mask, one_list, 3)

def draw_ICM(img, mask, label_json_object): #img, label_json['objects']
    for o in label_json_object:
        if o['class_name'] == 'ICM':
            icm_coord = o['annotation']['coord']['points']
            # print(len(icm_coord), icm_coord[0][0])
        
            icm_coords = icm_coord[0][0]
            # print(len(icm_coords))
            
            points = []
            for k in range(len(icm_coords)):
                points.append([int(icm_coords[k]['x']), int(icm_coords[k]['y'])])
            
            input_points = np.array(points, np.int32)
            # print(input_points)
            cv2.fillPoly(img, [input_points], (255,0,0))
            cv2.fillPoly(mask, [input_points], 1)
            
def draw_TE(img, mask, label_json_object):
    one_list = []
    for o in label_json_object:
        # print(o['class_name'])
        if o['class_name'] == 'TE':
            te_coord = o['annotation']['coord']['points'][0]
            # print(len(te_coord))

            all_input_points = []
            if len(te_coord) == 2:
                for te in te_coord:
                    points = []
                    for t in range(len(te)):
                        points.append([int(te[t]['x']), int(te[t]['y'])])
                    input_points = np.array(points, np.int32)
                    one_list.append(input_points)
            
            else:
                for te in te_coord[0]:
                    all_input_points.append([int(te['x']), int(te['y'])])
                    
                all_input_points = np.array(all_input_points, np.int32)
                one_list.append(all_input_points)

    cv2.fillPoly(img, one_list, (0,0,255))
    cv2.fillPoly(mask, one_list, 2)
```
